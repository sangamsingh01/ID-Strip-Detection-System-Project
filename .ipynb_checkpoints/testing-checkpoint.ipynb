# Prediction using model

```python
from tensorflow.keras import layers
from tensorflow.keras.models import load_model
import cv2
import matplotlib.pyplot as plt
from tensorflow.keras.optimizers import Adam  # Adjust the optimizer type accordingly

```

```python
model = load_model("detector.model")
print("model_loaded")
```

```python
test_img = cv2.imread(r'C:\Users\sanga\Documents\IDetection Project\wo.jpg')
plt.imshow(test_img)
```

```python
test_img.shape
test_img = cv2.resize(test_img,(400,1200))
test_input = test_img.reshape((1,400,1200,3))
```
